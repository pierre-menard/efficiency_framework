# Ethereum

```python
num_transactions = pd.read_csv("export-TxGrowth.csv")
num_transactions["Date(UTC)"] = pd.to_datetime(num_transactions["Date(UTC)"])
num_transactions.rename(columns = {"Value": "Number of Transactions"}, inplace = True)
num_transactions
```

```python
avg_gas = pd.read_csv("export-AvgGasPrice.csv")
avg_gas["Date(UTC)"] = pd.to_datetime(avg_gas["Date(UTC)"])
avg_gas.rename(columns = {"Value (Wei)": "Average Gas Price (Wei)"}, inplace=True)
avg_gas
```

```python
eth_price = pd.read_csv("export-EtherPrice.csv")
eth_price["Date(UTC)"] = pd.to_datetime(eth_price["Date(UTC)"])
eth_price.rename(columns = {"Value": "Ethereum Price"}, inplace = True)
eth_price
```

```python
eth_gas_price = pd.merge(avg_gas, eth_price, on = "Date(UTC)", how = "inner")
eth_gas_price["Average Gas (USD)"] = eth_gas_price["Average Gas Price (Wei)"] * eth_gas_price["Ethereum Price"] / 1_000_000_000_000_000_000
eth_gas_price
```

```python
eth_transaction_fess = pd.read_csv("export-AverageDailyTransactionFee.csv")
eth_transaction_fess["Date(UTC)"] = pd.to_datetime(eth_transaction_fess["Date(UTC)"])
eth_transaction_fess
```

```python
eth_costs_df = pd.merge(num_transactions, eth_gas_price, how = "inner", on = "Date(UTC)")
eth_costs_df = pd.merge(eth_costs_df, eth_transaction_fess, how = "inner", on = "Date(UTC)")
eth_costs_df = eth_costs_df[["Date(UTC)", "Number of Transactions", "Average Gas (USD)", "Average Txn Fee (USD)"]]
eth_costs_df["Total Cost (USD)"] = (eth_costs_df["Average Gas (USD)"] + eth_costs_df["Average Txn Fee (USD)"]) * eth_costs_df["Number of Transactions"]
eth_costs_df
```

```python
eth_total_volume = pd.read_csv("ITB_eth_transactions_volume_in_usd_ethereum_2023-12-05T21_30_10.982Z.csv")
eth_total_volume.rename(columns = {"DateTime": "Date(UTC)", "Total Volume": "Total Volume (USD)"}, inplace = True)
eth_total_volume["Date(UTC)"] = pd.to_datetime(eth_total_volume["Date(UTC)"])
eth_total_volume
```

```python
eth_percent_cost = pd.merge(eth_costs_df, eth_total_volume, on = "Date(UTC)", how = "inner")
eth_percent_cost["Percent Cost"] = eth_percent_cost["Total Cost (USD)"] / eth_percent_cost["Total Volume (USD)"]
eth_percent_cost
```

```python
print(eth_percent_cost[-365:]["Percent Cost"].mean() * 100)
```
